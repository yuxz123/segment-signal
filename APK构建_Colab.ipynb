```
# ============================================
# 第三步：下载APK
# ============================================
import os
from google.colab import files

apk_path = '/content/segment-signal/bin/segmentsignal-1.0-debug.apk'
if os.path.exists(apk_path):
    size_mb = os.path.getsize(apk_path) / 1024 / 1024
    print(f'APK大小: {size_mb:.1f} MB')
    print('正在触发浏览器下载...')
    files.download(apk_path)
else:
    print('[FAIL] 没找到APK，看上方第二步的红色报错')
    print('搜一下有没有其他apk文件:')
    !find /content/segment-signal -name '*.apk' 2>/dev/null
```
